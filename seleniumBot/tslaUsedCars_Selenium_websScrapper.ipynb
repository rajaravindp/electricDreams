```python
# Import requried libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd
import re
```

```python
class tslaBot():
    # Define the constructor
    def myFunc(self, model, year, zipCode):
        """ 
        Model = [ms, mx, m3, my]
        year = 2016 - 2022
        """
        # Define the path to the Chrome driver
        service = Service(executable_path='C://Users/Aravind/Downloads/chromedriver-win64/chromedriver-win64/chromedriver.exe')
        # Options for Chrome driver
        options = webdriver.ChromeOptions()
        # Create a new instance of the Chrome driver
        self.driver = webdriver.Chrome(service=service, options=options)

        # Navigate to tsla website
        website = f'https://www.tesla.com/inventory/used/{model}?Year={year}&arrangeby=plh&zip={zipCode}'
        # Get the website
        self.driver.get(website)
        # Maximise the window
        # self.driver.maximize_window()
        
        # Create an empty dataframe - specs
        # DAS - Driver Assistance System
        # Some issue scraping accident history

        specs = pd.DataFrame(columns=["model", "year", "odometer", "price", "location", "driveTrain", "DAS", "accident_history", "paintJob", "wheels", "emi", "zipCode"])

        # Find model name
        # model_index = ['Model S', 'Model X', 'Model 3', 'Model Y']
        model_name = self.driver.find_elements(By.XPATH, "(//span[contains(text(),'Model')])")
        # Extract the text and append to specs dataframe
        model_data = pd.Series(model_name).apply(lambda x: x.text)
        specs["model"] = model_data
        
        # Find model year
        model_year = self.driver.find_elements(By.XPATH, "(//span[contains(text(),'2018')])")
        # Extract the text
        year_data = pd.Series(model_year).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["year"] = year_data
        
        # Find odometer
        odometer = self.driver.find_elements(By.XPATH, "//div[contains(text(), 'odometer')]")
        # Extract the text
        odometer_data = pd.Series(odometer).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["odometer"] = odometer_data

        # Find price
        price = self.driver.find_elements(By.XPATH, "(//div[@class='result-price'])//span")
        # Extract the text
        price_data = pd.Series(price).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["price"] = price_data

        # Find location
        location = self.driver.find_elements(By.XPATH, "//article//section[1]//div[1]//div[3]")
        # Extract the text
        location_data = pd.Series(location).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["location"] = location_data

        # Find driveTrain
        driveTrain = self.driver.find_elements(By.XPATH, "//div[contains(text(), 'Drive')]")
        # Extract the text
        driveTrain_data = pd.Series(driveTrain).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["driveTrain"] = driveTrain_data
        
        # Find DAS
        driver_assistance = self.driver.find_elements(By.XPATH, "//article//section[5]//ul[1]//li[1]")
        # Extract the text
        driver_assistance_data = pd.Series(driver_assistance).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["DAS"] = driver_assistance_data

        # Find accident history
        accident_history = self.driver.find_elements(By.XPATH, "//article//section[5]//ul[1]//li[2]")
        # Extract the text
        accident_history_data = pd.Series(accident_history).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["accident_history"] = accident_history_data

        # Find paintJob
        paintJob = self.driver.find_elements(By.XPATH, "//article//section[5]//ul[2]//li[1]")
        # Extract the text
        paintJob_data = pd.Series(paintJob).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["paintJob"] = paintJob_data

        # Find wheels size
        wheels = self.driver.find_elements(By.XPATH, "//article//section[5]//ul[2]//li[2]")
        # Extract the text
        wheels_data = pd.Series(wheels).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["wheels"] = wheels_data

        # Find Interior
        interior = self.driver.find_elements(By.XPATH, "//article//section[5]//ul[2]//li[3]")
        # Extract the text
        interior_data = pd.Series(interior).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["interior"] = interior_data

        # Find EMI
        emi = self.driver.find_elements(By.XPATH, "//article//section[1]//div[2]//div[2]")
        # Extract the text
        emi_data = pd.Series(emi).apply(lambda x: x.text)
        # Append to specs dataframe
        specs["emi"] = emi_data

        # Find zipCode
        specs["zipCode"] = zipCode

        # Quit driver
        self.driver.quit()
        # Return specs
        return specs
```
